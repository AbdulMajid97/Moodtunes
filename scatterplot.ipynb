```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv("C:\\Users\\Majid\\Programming Files\\FYP\Datasets\\EMO_Dataset\\annotations\\static_annotations.csv")
```

```python
df
```

```python
plt.scatter(df['mean_arousal'], df['mean_valence'])
plt.xlabel('Mean Arousal')
plt.ylabel('Mean Valence')
plt.title('Scatter Plot of Mean Arousal vs Mean Valence')
```

```python
# Select the features for clustering
features = ['mean_arousal', 'mean_valence']
X = df[features]

# Apply K-means clustering
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
```

```python
df['cluster']= kmeans.labels_
```

```python
# Scatter plot with cluster colors
plt.scatter(df['mean_arousal'], df['mean_valence'], c=df['cluster'], cmap='viridis')
plt.xlabel('Mean Arousal')
plt.ylabel('Mean Valence')
plt.title('Clustering of Mean Arousal vs Mean Valence')

# Add data labels
for i in range(len(df)):
    plt.text(df['mean_arousal'][i], df['mean_valence'][i], df['song_id'][i])

# Show the plot
plt.show()
```
